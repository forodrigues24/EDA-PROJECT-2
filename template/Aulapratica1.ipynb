```python
# Instalar a biblioteca amply no Colab
!pip install -q amplpy

# Importar o AMPL
from amplpy import AMPL, ampl_notebook

# Inicializar AMPL no Colab
ampl = ampl_notebook(
    modules=["highs", "gurobi"],  # Usa "highs" (grátis) ou "gurobi" (licença)
)

# Definir variáveis no AMPL
ampl.eval("""
var xA >= 0;  # Meters of Fabric A
var xB >= 0;  # Meters of Fabric B

# Objective function (maximize revenue)
maximize Profit: 500 * xA + 200 * xB;

# Constraints
subject to Yellow_Wool:  400 * xA + 500 * xB <= 100000;
subject to Green_Wool:   500 * xA + 200 * xB <= 100000;
subject to Black_Wool:   300 * xA + 800 * xB <= 120000;

# Escolher solver
option solver highs;

# Solve the problem
solve;

# Display results
display xA, xB;
""")
```
